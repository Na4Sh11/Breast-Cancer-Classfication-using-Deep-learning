Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df.head()
```

```python
#adding target to the dataframe
df['label'] = dataset.target
```

```python
df.head()
```

```python
df.shape
```

```python
#checking missing values
df.info()
```

```python
#for statistics
df.describe()
```

```python
#distribution of benign and malignant
df['label'].value_counts()
```

1 -> Benign

0 -> Malignant

```python
df.groupby('label').mean()
```

Getting features and targets

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting train and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#standardize the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

Neural network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#layers of neural network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# train neural networks

history = model.fit(X_train_std, Y_train, validation_split = 0.1, epochs = 10)
```

```python
#visualizing accuracy

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Accuracy of the model')

plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
#visualizing accuraclossy

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Loss of the model')

plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
# accuracy of model for test data

loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
Y_pred = model.predict(X_test_std)
```

```python
print(Y_pred.shape)
print(Y_pred[0])
```

```python
#covert pred probability to label

Y_pred_label = [np.argmax(i) for i in Y_pred]
print(Y_pred_label)
```

```python
#building the predictive system

input_data = (13.53, 10.94, 87.91, 559.2, 0.1291, 0.1047, 0.06877, 0.06556, 0.2403, 0.06641, 0.4101, 1.014, 2.652, 32.65, 0.0134, 0.02839, 0.01162, 0.008239, 0.02572, 0.006164, 14.08, 12.49, 91.36, 605.5, 0.1451, 0.1379, 0.08539, 0.07407, 0.271, 0.07191
)

#change input to numpy array
input_data_as_np_array = np.asarray(input_data)

#reshape the np array
input_data_reshaped = input_data_as_np_array.reshape(1,-1)

#standardize the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(i) for i in prediction]
print(prediction_label)

if prediction_label[0] == 0:
  print("The tumor is Malignant")
else:
  print("The tumor is Benign")
```
